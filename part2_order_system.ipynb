# Task 1 — Explore the Menu
The input data provided for the menu is a nested dictionary. In order to print the full menu grouped by categories, I will need a list of unique categories in the menu. The question also hints at using a nested looping logic - looping through categories first and then looping through menu items and filtering by category. I can create a separate dataset of the categories first and then use it in another loop to identify and print menu items in that category. For the second part of the task, number of items on the menu is just the len() of the menu. For available items, I need to loop through and increment a counter for available items. To find the most expensive item, I will simply start at the top and keep comparing each subsequent item and replacing the variable with the more expensive item. A similar approach for the "items under 150", with a counter to increment for each item under 150 should work.

```python
# Input Data

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# Now to get the list of unique categories

categories = []
for item in menu:
    category = menu[item]["category"]
    if category not in categories:
        categories.append(category) # this will prevent duplicates in the categories list

for category in categories:
    print(f"--- {category} ---")
    for item in menu:
        if menu[item]["category"] == category:
            price = menu[item]["price"]
            available = menu[item]["available"]
            status = "Available" if available else "Unavailable"
            print(f"  - {item:<17} ₹{price:.2f}   [{status}]")
    print()  # printing a blank line between categories

print()
print("---- Menu Statistics ----")
print(f"Total items: {len(menu)}") # This should print the total number of items in the menu    

available_counter = 0
for item in menu:
    if menu[item]["available"]:
        available_counter += 1
print(f"Available items: {available_counter}")


expensive_name = ""
expensive_price = 0.0
for item in menu:
    if menu[item]["price"] > expensive_price:
        expensive_price = menu[item]["price"]
        expensive_name = item
print(f"Most expensive item: {expensive_name} at ₹{expensive_price}")

under150_counter = 0
for item in menu:
    if menu[item]["price"] < 150.0:
        under150_counter += 1
print(f"Items under ₹150: {under150_counter}")

if available_counter > 0 and under150_counter > 0:
    print("Items under ₹150:")
    for item in menu:
        if menu[item]["price"] < 150.0:
            print(f"  - {item} at ₹{menu[item]['price']:.2f}")
else:
    print("No items under 150.")
        
```

# Task 2 — Cart Operations
Starting with a blank cart, I need to be able to simulate a "add to cart" functionality. I will need to write a function that I can call for each addition to the cart. I will first check if the item exists on the menu and is available. If not, then I will simply message appropriately. Else, I need to check if it already is in the cart and increment the count OR add it to the cart. Similarly, I will have to write a function to remove items from the cart. Need to check if it exists in the cart and decrement quantity if more than 1 and item itself if quantity = 1. I also need a function to update the quantity in the cart (add or delete)

```python
cart = []
def add_to_cart(item_name, quantity):
    print(f"Adding {quantity} of {item_name} to cart...")


    if item_name not in menu:
        print(f"Item '{item_name}' not found in menu.")
        return  
        
    if not menu[item_name]["available"]:
        print(f"Item '{item_name}' is not available.")
        return
    
    # If we reach here, the item is valid and available. Now to add it to the cart.
    item_found_in_cart = False
    for cart_item in cart:
        if cart_item["name"] == item_name:
            cart_item["quantity"] += quantity
            print(f"Updated {item_name} quantity in cart to {cart_item['quantity']}.")
            item_found_in_cart = True
            break
    if not item_found_in_cart:
        cart.append({"name": item_name, "quantity": quantity})
        print(f"Added {item_name} to cart with quantity {quantity}.")

def remove_from_cart(item_name):
    print(f"Removing {item_name} from cart...")
    for cart_item in cart:
        if cart_item["name"] == item_name and cart_item["quantity"] == 1:
            cart.remove(cart_item)
            print(f"Removed {item_name} from cart.")
            return
        else:
            update_cart(item_name, cart_item["quantity"] - 1)
            print(f"Decreased {item_name} quantity in cart to {cart_item['quantity']}.")
            return
    print(f"Item '{item_name}' not found in cart.")

def update_cart(item_name, new_quantity):
    print(f"Updating {item_name} quantity to {new_quantity} in cart...")
    if new_quantity <= 0:
        print(f"   Quantity is {new_quantity}, removing item instead.")
        remove_from_cart(item_name)
        return

    for cart_item in cart:
        if cart_item["name"] == item_name:
            cart_item["quantity"] = new_quantity
            print(f"Updated {item_name} quantity in cart to {new_quantity}.")
            return
    print(f"Item '{item_name}' not found in cart.")

def print_cart():
    print("Current Cart:")
    if not cart:
        print("  Cart is empty.")
        return
    for cart_item in cart:
        print(f"  - {cart_item['name']}: {cart_item['quantity']}")

# Now ready to simulate cart operations as per question
print("\n--- Simulating Cart Operations ---")

# 1. Add Paneer Tikka X 2 to the cart
add_to_cart("Paneer Tikka", 2)
print_cart()

# 2. Add Gulab Jamun X 1 to the cart
add_to_cart("Gulab Jamun", 1)
print_cart()

# 3. Add Paneer Tikka X 1 to the cart
add_to_cart("Paneer Tikka", 1)
print_cart()

# 4. Add Mystery Burger X 1 to the cart (which does not exist)
add_to_cart("Mystery Burger", 1)
print_cart() # the cart should remain unchanged

# 5. Add Chicken Wings X 1 to the cart (which is unavailable)
add_to_cart("Chicken Wings", 1)
print_cart() # the cart should remain unchanged

# 6. Remove Gulab Jamun from the cart
remove_from_cart("Gulab Jamun")
print_cart()

```

# Task 3 - Inventory Tracker with Deep Copy
Need to use deepcopy() in copy package for this. I will need to deepcopy the inventory dataset to a backup dataset and then remove items in my cart (above task) to demonstrate that backup is not affected. I also need to ensure that I cannot remove more quantity that what already exists in inventory levels and give a warning. Then restore inventory from backup. Also, I need to loop through with each deduction and print a reorder alert if quantity falls below the reorder_level. Finally need to print the inventory and the backup to confirm that they differ.

```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  4, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# Items left in the cart from previous task
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0},
    {"item": "Gulab Jamun", "quantity": 2, "price": 90.0},
]

print("\n--- Demonstrating Deep Copy ---")
inventory_backup = copy.deepcopy(inventory)
print("Sample item in Original inventory & Backup before modification:")
print(f"  Gulab Jamun stock (original)       = {inventory['Gulab Jamun']['stock']}")
print(f"  Gulab Jamun stock (backup)         = {inventory_backup['Gulab Jamun']['stock']}")

inventory["Gulab Jamun"]["stock"] -= 2
print("\nAfter modifying original inventory:")
print(f"  Gulab Jamun stock (original)       = {inventory['Gulab Jamun']['stock']}")
print(f"  Gulab Jamun stock (backup)         = {inventory_backup['Gulab Jamun']['stock']}")

# Restore original inventory from backup
inventory = copy.deepcopy(inventory_backup)
print("\nAfter restoring from backup:")
print(f"  Gulab Jamun stock (original)       = {inventory['Gulab Jamun']['stock']}")
print(f"  Gulab Jamun stock (backup)         = {inventory_backup['Gulab Jamun']['stock']}")

# Now to simulate the order processing and deduction from inventory
print("\n--- Simulating Order Processing and Inventory Update ---")
for cart_item in cart:
    item_name = cart_item["item"]
    quantity = cart_item["quantity"]
    current_stock = inventory[item_name]["stock"] # I need this to check if we have enough stock before processing the order
    if item_name in inventory:
        if current_stock >= quantity:
            inventory[item_name]["stock"] -= quantity
            print(f"Processed {quantity} of {item_name}. Remaining stock: {inventory[item_name]['stock']}")
        else:
            print(f"Insufficient stock for {item_name}. Available: {current_stock}, Requested: {quantity}")
            print(f"   Deducting {current_stock} instead of {quantity}.")
            inventory[item_name]["stock"] = 0
    else:
        print(f"Item '{item_name}' not found in inventory.")

# Setting up Reorder Alerts
print("\n--- Reorder Alerts ---")
for item in inventory:
    stock = inventory[item]["stock"]
    reorder_level = inventory[item]["reorder_level"]
    alert_count = 0
    if stock <= reorder_level:
        print(f"Reorder Alert: {item} stock is low (Current: {stock}, Reorder Level: {reorder_level})")
        alert_count = alert_count + 1
    
    if alert_count == 0:
        print(f"{item} stock is sufficient (Current: {stock}, Reorder Level: {reorder_level})")
    else:        print(f"{item} needs to be reordered. Current stock: {stock}")

# Differences between Inventory and Inventory Backup after processing the order
print("\n--- Inventory vs Inventory Backup ---")
for item in inventory:
    current_stock = inventory[item]["stock"]
    backup_stock = inventory_backup[item]["stock"]
    if current_stock != backup_stock:
        print(f"{item}: Current Stock = {current_stock}, Backup Stock = {backup_stock}, Difference = {backup_stock - current_stock}")
    else:        
        print(f"{item}: Stock is unchanged at {current_stock}.")


```

# Task 4 - Daily Sales Log Analysis
I have the sales_data as input. Need to write a function that will loop through and aggregate sales values for the same date. Also, need to loop through and compare revenues to find best revenue and corresponding date. Need to loop through orders to increment an order count variable for each item, to find the most ordered item. After adding the new entry, I can reuse the same function to print the new updated log.

```python
# Input Data

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

def print_sales_log(sales_log):
    best_date = ""
    best_revenue = 0.0

    for date in sales_log:
        day_total = 0
        for order in sales_log[date]:
            day_total += order["total"]
        
        order_count = len(sales_log[date])
        print(f"  {date}  |  {order_count} order(s)  |  ₹{day_total:.2f}")
        if day_total > best_revenue:
            best_revenue = day_total
            best_date = date
    return best_date, best_revenue

# 1. Revenue per day
best_date, best_revenue = print_sales_log(sales_log)

# 2. Best-selling day
print(f"\nBest day: {best_date} with revenue of ₹{best_revenue:.2f}")

# 3. Most ordered Item
item_order_count = {}
for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_order_count:
                item_order_count[item] = 0
            item_order_count[item] += 1

most_ordered_item = max(item_order_count, key=item_order_count.get)
print(f"\nMost ordered item: {most_ordered_item} with {item_order_count[most_ordered_item]} orders")

# 4. Adding a new day to the sales log and reprinting the updated log and best day

# Input data for the new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("\n New day added. Updated Sales Log:")
best_date, best_revenue = print_sales_log(sales_log)
print(f"\nBest day: {best_date} with revenue of ₹{best_revenue:.2f}")

# 5. Use enumerate to print numbered list of all orders across all days
print("\n--- All Orders with Order IDs ---")
all_orders = []

for date in sales_log:
    for order in sales_log[date]:
        all_orders.append((date, order))

for i, (date, order) in enumerate(all_orders, start=1):
    oid = order["order_id"]
    total = order["total"]
    items_str = ", ".join(order["items"])
    print(f"  {i}. [{date}] Order #: {order['order_id']}, Total: ₹{total:.2f}, Items: {items_str}")
```
